```python
import sys
from pandas import DataFrame   #DataFrame通常来装二维的表格
import pandas as pd            #pandas是流行的做数据分析的包

#建立字典，键和值都从文件里读出来。
dict_data1={}
dict_data2={}
dict_data3={}

#打开文件
#此处处理的是日频raw data，若要处理分钟频raw data，将下面的dailyData改为minuteData即可
with open('E:\myToyData\dailyData1\part-r-00000','r')as df1:
    #读每一行
    for line in df1:
        #如果这行是换行符就跳过，这里用'\n'的长度来找空行
        if line.count('\n') == len(line):
            continue
        #对每行清除前后空格（如果有的话），然后用"："分割
        for kv in [line.strip().split('\t')]:
            #按照键，把值写进去
            dict_data1.setdefault(kv[0],[]).append(kv[1])

with open('E:\myToyData\dailyData2\part-r-00000','r')as df2:
    #读每一行
    for line in df2:
        #如果这行是换行符就跳过，这里用'\n'的长度来找空行
        if line.count('\n') == len(line):
            continue
        #对每行清除前后空格（如果有的话），然后用"："分割
        for kv in [line.strip().split('\t')]:
            #按照键，把值写进去
            dict_data2.setdefault(kv[0],[]).append(kv[1])

with open('E:\myToyData\dailyData3\part-r-00000','r')as df3:
    #读每一行
    for line in df3:
        #如果这行是换行符就跳过，这里用'\n'的长度来找空行
        if line.count('\n') == len(line):
            continue
        #对每行清除前后空格（如果有的话），然后用"："分割
        for kv in [line.strip().split('\t')]:
            #按照键，把值写进去
            dict_data3.setdefault(kv[0],[]).append(kv[1])
            
            #print（dict_data）看看效果
```

```python
df1 = pd.DataFrame.from_dict(dict_data1, orient='index',columns=['mixed'])
df1 = df1.reset_index().rename(columns = {'index':'time'})
df2 = pd.DataFrame.from_dict(dict_data2, orient='index',columns=['mixed'])
df2 = df2.reset_index().rename(columns = {'index':'time'})
df3 = pd.DataFrame.from_dict(dict_data3, orient='index',columns=['mixed'])
df3 = df3.reset_index().rename(columns = {'index':'time'})
df=pd.concat([df1,df2,df3],axis=0)
df
```

```python
df_split=pd.concat([df['time'].str.split('_',expand=True),df['mixed'].str.split(',',expand=True)],axis=1)
df_split.columns=['stock','time','open','high','low','close','volumn','amount']
#df_split.rename(columns={'0':'open','1':'high','2':'low','3':'close','4':'volumn','5':'amount'}, inplace=False)
df_split
```

```python
factor_list = ['open','high','low','close','volumn','amount']
for factor in factor_list:
    df_temp=df_split[['time','stock',factor]]
    s = pd.Series(list(df_temp[factor]),index=[df_temp['time'],df_temp['stock']])  # 利用一维数据表df2构造一个二级索引Series对象
    new_df = s.unstack() # 对s执行stack的逆操作unstack
    outputpath='E:/myToyData/output/day_'+factor+'.csv'
    new_df.to_csv(outputpath,sep=',',index=True,header=True)
```
